````python
# Funciton for visualizing confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import itertools

def plot_confusion_matrix(cm, classes, title='Confusion matrix',
                          cmap=sns.cubehelix_palette(as_cmap=True)):
    """
    This function is modified from: 
    http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
    """
    classes.sort()
    tick_marks = np.arange(len(classes))    
    
    fig, ax = plt.subplots(figsize=(5,5))
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           xticklabels = classes,
           yticklabels = classes,
           title = title,
           xlabel = 'Predicted label',
           ylabel = 'True label')

    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")
    ylim_top = len(classes) - 0.5
    plt.ylim([ylim_top, -.5])
    plt.tight_layout()
    plt.show()
````

````python
# plot your confusion matrix
my_tags = ['anger', 'fear', 'joy', 'sadness']
plot_confusion_matrix(cm, classes=my_tags, title='Confusion matrix')
````

````python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import LabelEncoder
import numpy as np


X_train = torch.tensor(X_train.toarray(), dtype=torch.float32)
X_test = torch.tensor(X_test.toarray(), dtype=torch.float32)

label_encoder = LabelEncoder()
y_train = torch.tensor(label_encoder.fit_transform(train_df['emotion']), dtype=torch.long)
y_test = torch.tensor(label_encoder.transform(test_df['emotion']), dtype=torch.long)

input_shape = X_train.shape[1]
output_shape = len(label_encoder.classes_)
print('input_shape:', input_shape)
print('output_shape:', output_shape)

batch_size = 32
train_dataset = TensorDataset(X_train, y_train)
test_dataset = TensorDataset(X_test, y_test)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

class SimpleNN(nn.Module):
    def __init__(self, input_size, output_size):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, 128)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(128, output_size)
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return self.softmax(x)

model = SimpleNN(input_shape, output_shape)

criterion = nn.CrossEntropyLoss()  
optimizer = optim.Adam(model.parameters(), lr=0.001)

epochs = 25
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for X_batch, y_batch in train_loader:

        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)


        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

      
        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += y_batch.size(0)
        correct += (predicted == y_batch).sum().item()

    print(f"Epoch [{epoch+1}/{epochs}], Loss: {running_loss/len(train_loader):.4f}, Accuracy: {correct/total:.4f}")


model.eval()
correct = 0
total = 0
with torch.no_grad():
    for X_batch, y_batch in test_loader:
        outputs = model(X_batch)
        _, predicted = torch.max(outputs, 1)
        total += y_batch.size(0)
        correct += (predicted == y_batch).sum().item()

print(f"Test Accuracy: {correct/total:.4f}")
````

````python
# Answer here
import ollama
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import WebBaseLoader
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import OllamaEmbeddings
from IPython.display import display, Markdown

llm_model = "llama3.2" 

def load_and_retrieve_docs(urls):
    all_retrievers = []
    for url in urls:
        loader = WebBaseLoader(
            web_paths=(url,),
            bs_kwargs=dict() 
        )
        docs = loader.load()
        text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)  
        splits = text_splitter.split_documents(docs)
        embeddings = OllamaEmbeddings(model=llm_model)  
        vectorstore = Chroma.from_documents(documents=splits, embedding=embeddings)  
        all_retrievers.append(vectorstore.as_retriever())
    return all_retrievers


urls = [
    "https://www.ibm.com/topics/large-language-models",
    "https://www.microsoft.com/en-us/ai/what-is-large-language-model",
    "https://aws.amazon.com/blogs/machine-learning/large-language-models/"
]


retrievers = load_and_retrieve_docs(urls)


def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

def ollama_llm(question, context):
    formatted_prompt = f"Question: {question}\n\nContext: {context}"
    response = ollama.chat(model=llm_model, messages=[{'role': 'user', 'content': formatted_prompt}])
    return response['message']['content']

def rag_chain(question, retrievers):
    combined_context = ""
    for retriever in retrievers:
        retrieved_docs = retriever.invoke(question)
        formatted_context = format_docs(retrieved_docs)
        combined_context += formatted_context + "\n\n"  
    return ollama_llm(question, combined_context)

questions = [
    "What are the related solutions of IBM with LLMs?",
    "What does Microsoft describe about the applications of LLMs?",
    "How does AWS explain the advantages of LLMs?"
]

for i, question in enumerate(questions):
    result = rag_chain(question, retrievers)
    print(f"Question {i+1}: {question}")
    display(Markdown(result))
    print("\n")
````

````python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


X_train_tensor = torch.tensor(np.vstack(df_train["embeddings"].to_numpy()), dtype=torch.float32)
X_test_tensor = torch.tensor(np.vstack(df_test["embeddings"].to_numpy()), dtype=torch.float32)
y_train_tensor = torch.tensor(df_train['emotion'].factorize()[0], dtype=torch.long)
y_test_tensor = torch.tensor(df_test['emotion'].factorize()[0], dtype=torch.long)


train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)


class SimpleNN(nn.Module):
    def __init__(self, input_size, output_size):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, 128)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(0.3)  
        self.fc2 = nn.Linear(128, 64)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(0.3)
        self.fc3 = nn.Linear(64, output_size)
    
    def forward(self, x):
        x = self.fc1(x)
        x = self.relu1(x)
        x = self.dropout1(x)
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.dropout2(x)
        x = self.fc3(x)
        return x  

input_size = X_train_tensor.shape[1]
output_size = len(df_train['emotion'].unique())
model = SimpleNN(input_size, output_size)


criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.5)  


log_data = {"epoch": [], "loss": [], "val_loss": [], "accuracy": [], "val_accuracy": []}


epochs = 50
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for X_batch, y_batch in train_loader:
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += y_batch.size(0)
        correct += (predicted == y_batch).sum().item()

    train_loss = running_loss / len(train_loader)
    train_accuracy = correct / total

    model.eval()
    val_loss = 0.0
    val_correct = 0
    val_total = 0
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)

            val_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            val_total += y_batch.size(0)
            val_correct += (predicted == y_batch).sum().item()

    val_loss /= len(test_loader)
    val_accuracy = val_correct / val_total

    log_data["epoch"].append(epoch + 1)
    log_data["loss"].append(train_loss)
    log_data["val_loss"].append(val_loss)
    log_data["accuracy"].append(train_accuracy)
    log_data["val_accuracy"].append(val_accuracy)

    scheduler.step()  

    print(f"Epoch [{epoch + 1}/{epochs}], Loss: {train_loss:.4f}, Accuracy: {train_accuracy:.4f}, "
          f"Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}")


log_df = pd.DataFrame(log_data)
log_df.to_csv("logs/training_log_improved.csv", index=False)
print("Training finished and log saved.")


model.eval()
y_test_pred = []
y_test_true = []
with torch.no_grad():
    for X_batch, y_batch in test_loader:
        outputs = model(X_batch)
        _, predicted = torch.max(outputs, 1)
        y_test_pred.extend(predicted.cpu().numpy())
        y_test_true.extend(y_batch.cpu().numpy())


accuracy = accuracy_score(y_test_true, y_test_pred)
print(f"Neural Network Test Accuracy: {accuracy:.4f}")


print("\nClassification Report:")
print(classification_report(y_test_true, y_test_pred))


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(log_data["epoch"], log_data["loss"], label="Training Loss")
plt.plot(log_data["epoch"], log_data["val_loss"], label="Validation Loss")
plt.title("Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(log_data["epoch"], log_data["accuracy"], label="Training Accuracy")
plt.plot(log_data["epoch"], log_data["val_accuracy"], label="Validation Accuracy")
plt.title("Accuracy Curve")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.tight_layout()
plt.show()

cm = confusion_matrix(y_test_true, y_test_pred)
my_tags = ['anger', 'fear', 'joy', 'sadness']
plot_confusion_matrix(cm, classes=my_tags, title='Confusion Matrix for Neural Network')
````

Compare and discuss the results of the zero-shot, 1-shot and 5-shot classification.

# Answer here
Zero-Shot Classification：
描述：模型在未見過任何標註樣本的情況下進行分類。
表現：準確率通常較低，因為缺乏任務相關的上下文。
原因：完全依賴預訓練知識，無法針對具體任務進行調整。
1-Shot Classification：
描述：模型每個類別只看到一個標註樣本。
表現：比零樣本有適度提升，但提升有限。
原因：單個樣本提供了一些任務上下文，但不足以讓模型充分泛化。
5-Shot Classification：
描述：模型每個類別看到五個標註樣本。
表現：比零樣本和單樣本分類的準確率顯著提高。
原因：更多的樣本為模型提供更清晰的任務特徵，提升了泛化能力。

為什麼結果不同
上下文與任務匹配：零樣本完全依賴預訓練知識，而單樣本和五樣本能提供部分或更多的任務相關信息。
數據效率：少樣本學習能讓模型以極少的標註數據快速適應任務。
模型行為：
零樣本：適合與預訓練知識高度相關的任務。
單樣本：在數據有限的情況下能提供初步改進。
五樣本：接近傳統監督學習的效果，同時所需數據量較少。
少樣本學習（1-shot & 5-shot）：
優勢：能融入任務特定的樣本；泛化能力更強。
挑戰：需要一定數量的標註數據，但需求較低。
